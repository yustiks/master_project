```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("result_resnet1.csv", header=None) 
```

```python
df = df[np.isfinite(df[11])]
```

```python
fig = plt.figure(figsize=(15,4))
plt.subplot(1, 2, 1)
x = np.linspace(1, 100, 100)
y = df.loc[:,5]
y1 = df.loc[:,11]
plt.plot(x, y, '-', x,y1,'-')
plt.legend(['training loss','validation loss'])
plt.ylim(0,0.8)
plt.xlabel('# step of epochs')
plt.ylabel('loss')


plt.subplot(1, 2, 2)
y = df.loc[:,8]
y1 = df.loc[:,14]
plt.plot(x, y, '-', x,y1,'-')
plt.legend(['training accuracy','validation accuracy'])
plt.ylim(0.75,1)
plt.xlabel('# step of epochs')
plt.ylabel('loss')
plt.savefig('resnet_dataset1.png')
plt.show()
```
